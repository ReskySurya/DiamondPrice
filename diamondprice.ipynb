```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from wordcloud import WordCloud
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("diamonds.csv")
data.head()
```

```python
data.describe()
```

```python
plt.scatter(data['depth'], data['price'])
plt.xlabel('depth')
plt.ylabel('price')
plt.show()
```

```python
#data = pd.read_csv("CarPrice.csv")
plt.figure(figsize=(30,4))

plt.subplot(1,2,1)
plt.title('Diamond Prize')
sns.histplot(data.price)

plt.show()
```

```python
x = data[['carat','depth','table']]
y = data['price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model_regresi = LinearRegression()
model_regresi.fit(X_train, y_train)

model_regresi_pred = model_regresi.predict(X_test)

plt.scatter(X_test.iloc[:, 0], y_test, label='Actual Prices', color='blue')
plt.scatter(X_test.iloc[:, 0], model_regresi_pred, label='Predicted Prices', color='red')
plt.xlabel('carat')
plt.ylabel('prices')
plt.legend()
plt.show()
```

```python
sns.violinplot(
    x='cut',
    y='price',
    data=data,
    palette='Set1',
    linewidth=2,
    alpha=0.7
)

# Add a title
plt.title('Violin Plot for Cut vs Price')

# Add labels to the axes
plt.xlabel('Cut')
plt.ylabel('Price')

# Show the plot
plt.show()
```

```python
carat = float(input('Masukan Nilai Carat = '))
depth = float(input('Masukan Nilai Depth = '))
table = float(input('Masukan Nilai Table = '))
X = np.array([[carat, depth, table]])
price_X = model_regresi.predict(X)

#mengubah nilai array menjadi interger
price_X_integer = price_X.astype(int)
print(f'Harga Prediksi Diamond adalah ${price_X_integer[0]}')
```
